```python
## find the duplicate eliments in this list

num_list = [10, 20, 10, 30, 40, 20, 50]

dup_values = [] ##10, 20, 
unique_values = []

for num  in num_list:
    if num_list.count(num) > 1:
        if num not in dup_values:
            dup_values.append(num)
    else:
        unique_values.append(num)

print("Duplicate values:", dup_values)
print("Unique values:", unique_values)
```

```python
## to check given string anagram

value1 = "listen"
value2 = "silent"

freq1 = {}
freq2 = {}

count1 = 0
count2 = 0


for char in value1:
    if char in freq1:
        freq1[char] += 1
    else:
        freq1[char] = 1

for char in value2:
    if char in freq2:
        freq2[char] += 1
    else:
        freq2[char] = 1

print(freq1)
print(freq2)

if freq1 == freq2:
    print("Anagram")
    print(True)
else:
    print("Not Anagram")
    print(False)
```
