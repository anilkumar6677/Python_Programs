```python
## find the duplicate eliments in this list

num_list = [10, 20, 30, 40, 20, 50]

dup_values = [] ##10, 20, 
unique_values = []

for num  in num_list:
    if num_list.count(num) > 1:
        if num not in dup_values:
            dup_values.append(num)
    else:
        unique_values.append(num)

print("Duplicate values:", dup_values)
print("Unique values:", unique_values)
```

```python
##find 2nd higest number in list ??

nums = [10, 5, 20, 8,8, 20,15,30]
n = len(nums)
for i in range(n):
    for j in range(n - 1):
        if nums[j] > nums[j+1]:
            nums[j],nums[j+1] = nums[j+1], nums[j]

print(nums, nums[-2])
print("second higest", list(set(nums))[-3])
```

```python
## what is custome exception in python 

"""  
Custome exceptiuon are exception which we need to raise the exception in our code accroding to our requirement


for this we need to create exception class and later onjy;l rasie the exception 
"""

class InsufficientBalance(Exception):
    pass


amount = 400
balance = 900
try:
    if amount > balance:
        raise InsufficientBalance("in your account balance is low")

except InsufficientBalance as e:
    print("insufficient balance", e)
else:
    print("no error in this code")

```
