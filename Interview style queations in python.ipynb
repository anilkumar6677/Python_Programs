```python
##2ï¸âƒ£ Remove duplicates (keep order)

##input   [1, 2, 2, 3, 1, 4]
##output [1, 2, 3, 4]

data = [1, 2, 2, 3, 1, 4]

dup_val_remove = []

for num in data:
    if num not in dup_val_remove:
        dup_val_remove.append(num)

print(dup_val_remove)
```

```python
##4ï¸âƒ£ Flatten nested list
##input  [1, [2, 3], [4, [5, 6]], 7]
#output  [1, 2, 3, 4, 5, 6, 7]

data = [1, [2, 3], [4, [5, 6]], 7]
result = []

def flat_data(lst):
    for num in lst:
        if isinstance(num, list):
            flat_data(num)
        else:
            result.append(num)

flat_data(data)
print(result)
    
        

```

```python
##5ï¸âƒ£ Count frequency of elements
## input [1, 2, 2, 3, 3, 3]
## output {1:1, 2:2, 3:3}


data = [1, 2, 2, 3, 3, 3]

freq = {}
for num in data:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)

# for key,value in freq.items():
#     print(key, value)
```
